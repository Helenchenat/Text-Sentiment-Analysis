```python
#Extracting contents from url
```

```python
from newspaper import Article
```

```python
#Extracting the text for the 1st row
```

```python
def text(url):
    
    article=Article(url)
    article.download()
    article.parse()
    article.nlp()
    
    return article.text

for i in range(0,len(data)):
    data['URL_ID'][i]=text(data['URL'][i])
```

```python
#Text preprocessing
```

```python
import nltk
```

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
def rem_stopwords(text):
    dat=re.sub('[^a-zA-Z0-9]', ' ',text)
    stop_words=set(stopwords.words('english'))
    dat=word_tokenize(text)
    filtered=[word for word in dat if word.lower() not in stop_words]
    return filtered

data['words']=data['URL_ID'].apply(rem_stopwords)
    
    
    
```

```python
data.head()
```

```python
#Finding word count
```

```python
data['word_count']=data['words'].apply(lambda x:len(str(x).split()))
```

```python
data.head()
```

```python
#Finding average no:of words per sentence
```

```python
from nltk import sent_tokenize
```

```python
def words_sent(text):
    sentences=sent_tokenize(text)
    total_words=0
    for sentence in sentences:
        total_words+=len(word_tokenize(sentence))
    average_word=total_words/len(sentences)
    return average_word
data['average no:of words per sentence']=data['URL_ID'].apply(words_sent)

    

```

```python
data.head()
    
```

```python
# Finding average word length
```

```python
def word_len(text):
    words=word_tokenize(text)
    total_char=0
    for word in words:
        total_char+=len(word)
    avg_len=total_char/len(words)
    return avg_len
data['Average word length']=data['URL_ID'].apply(word_len)
        
    
```

```python
data.head()
```

```python
# find number of syllables
```

```python
# Here pyphen library is used to find the number of syllables
```

```python
#pip install pyphen
```

```python
import pyphen
dic=pyphen.Pyphen(lang='en')
    
```

```python
def syllable_count(text):
    words = text.split()
    total_syllables = 0
    for word in words:
        hyphenated_word = dic.inserted(word)
        total_syllables += len(hyphenated_word.split('-'))
    return total_syllables
data['syllables']=data['URL_ID'].apply(syllable_count)
```

```python
data.head()
```

```python
#Finding complex words(Those words having more than 1 syllable)
```

```python
def complex_words(text):
    words = text.split()
    complex_word=0
    for word in words:
        hyphenated_word = dic.inserted(word)
        if len(hyphenated_word.split('-'))>1:
            complex_word+=1  
    return complex_word
data['complex words']=data['URL_ID'].apply(complex_words)
```

```python
data.head()
```

# Sentiment Analysis

```python
#Cleaning using stopword
```

```python
import os
def stopword_clean(text,stopword_folder):
    stop_words=set()
    for filename in os.listdir(stopword_folder):
        with open(os.path.join(stopword_folder, filename), 'r') as file:
            stop_words.update(set(file.read().split()))
            
    words=text.split()
    clean_words=[word for word in words if word.lower() not in stop_words]
    return " ".join(clean_words)


data['clean_words']=data['URL_ID'].apply(lambda x:stopword_clean(x,"StopWords"))

```

```python
data.head()
```

```python
#Finding positive score
```

```python
def pos_score(text,Masterdictionary):
    positive_word=set()
    with open(os.path.join(Masterdictionary,"positive-words.txt"),'r') as f1:
            positive_word.update(set(f1.read().split()))
    words=text.split()
    words=[word.lower() for word in words]
    positive_score=0
    for word in words:
        if word in positive_word:
            positive_score+=1
            
    return positive_score
data["positive_score"]=data['clean_words'].apply(lambda x:pos_score(x,"MasterDictionary"))
            
    
       
        
    
```

```python
data.head()
```

```python
def neg_score(text,Masterdictionary):
    negative_word=set()
    with open(os.path.join(Masterdictionary,"negative-words.txt"),'r') as f1:
            negative_word.update(set(f1.read().split()))
    words=text.split()
    words=[word.lower() for word in words]
    negative_score=0
    for word in words:
        if word in negative_word:
            negative_score+=1
            
    return negative_score
data["negative_score"]=data['clean_words'].apply(lambda x:neg_score(x,"MasterDictionary"))
```

```python
data.head()
```

```python
#Finding Polarity
```

```python
data["Polarity_Score"]=data["positive_score"]-data["negative_score"]
```

```python
data.head()
```

```python
#Finding Subjectivity score
```

```python
def clean_words(text):
    words=text.split()
    count=len(words)
    return count
data['count']=data['clean_words'].apply(clean_words)

#word_count=len(data["clean_words"].split())

data['Subjectivity_score']=((data['positive_score']+data['negative_score'])/data['count'])+.000001
```
